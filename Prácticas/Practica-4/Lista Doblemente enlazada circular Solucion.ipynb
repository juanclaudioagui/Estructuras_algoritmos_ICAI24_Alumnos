```python
class Nodo:
    def __init__(self, valor=None):
        self.valor = valor
        self.next = None
        self.prev = None
        
    def __str__(self):
        str=(f'\nEl valor del nodo es : {self.num}')        
        return str

class ListaEnlazada:
    def __init__(self):
        self.head = None
        self.tail = None
    
    
    #  Insertion Method in Doubly Linked List
    def AñadirNodo(self, nodo): #Indicamos si queremos insertar al principio o al final.
        if self.head is None:
            self.head=nodo   
            self.tail=nodo 
        else: 
         
                nodo.next = None
                nodo.prev = self.tail
                self.tail.next = nodo
                self.tail = nodo
            
    
    #  Recorrido desde la cabeza de lista
    def RecorrerListaCabeza(self):
        print("RECORRIDO DE LISTA DESDE LA CABEZA: ")
        if self.head is None:
            print("No existen elementos para recorrer")
        else:
            aux = self.head
            while aux:
                print(aux.valor)
                aux = aux.next
    
    #  Recorrido desde la cola de lista
    def RecorrerListaCola(self):
        print("RECORRIDO DE LISTA DESDE LA COLA: ")
        if self.head is None:
            print("No existen elementos para recorrer")
        else:
            aux = self.tail
            while (aux!=None):
                print(aux.valor)
                aux = aux.prev    

def Menu():
    print("MENU");
    print("1. Insertar un nuevo valor desde la cola de lista.")
    print("2. Mostrar todos los nodos de la lista desde la cabecera. ")   
    print("3. Mostrar todos los nodos de la lista desde la cola en orden inverso. ")  
    print("4. Salir.")
    opc=int(input("Introduzca una opcion: "))
    return opc           

if __name__ == "__main__":
    lista=ListaEnlazada()
    opc=Menu()
    while(opc!=4):
        if (opc==1):
            nodo=Nodo()   
            valor=int(input("Intrdozuca el valor del nodo: ")) 
            nodo.valor=valor                    
            lista.AñadirNodo(nodo)
        elif(opc==2):
            lista.RecorrerListaCabeza()
        elif(opc==3):
            lista.RecorrerListaCola()
            
        elif(opc==4):
            print("Gracias por utilizar software ICAI")
     
        
            
        opc=Menu()
```
