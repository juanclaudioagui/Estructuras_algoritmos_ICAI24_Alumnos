# Iterative algoritm for Zero of polinomial
The newton-raphson alg is a simple "descent" algoritm  for finding a solution of an qautions

lets take the following equatoin

$$ x^3 + 6 x^2 + 3 x -10 = 0 $$

Lets plot it...


```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np

print ("Libs random, math and pyplot loaded")
```

Lets plot it

```python
def f(x): return  x**3 + 6* x**2 + 3*x -10
def fp(x): return 3*x**2 + 12*x + 3
```

```python

x = np.arange(-3,3,0.01,dtype=float)
y = [f(xval) for xval in x ]

fig, ax = plt.subplots(figsize=(10,5))

ax.set(xlabel='x', ylabel='y',title=f"Function to find roots")
ax.plot(x,y,linewidth=1)
ax.hlines([0],-3,3)
ax.vlines([0],1.5*min(y),max(y))
xr =[-2,1]
yr=[0,0]
ax.plot(xr,yr,marker='o',color='b')

ax.grid()
plt.show()
    
```

Lets see how it iterates towards the root, starting at say, x = 3

$$ x_{n+1} = x_n - {f(x_n)\over{f^{\prime}(x_n)}} $$ 

```python
# now iterate
epsilon = 1E-20
x= -0.25
error = abs(f(x)-0.0)
iter = 0 
xp=[x]
ep=[error]
while error > epsilon:
    print (f"{iter:2d}\t {x:.10f} \t {f(x):.10f}")
    x = x - f(x)/fp(x)
    error = abs(f(x) - 0.0)
    xp.append(x)
    ep.append(error)
    iter += 1
print (iter, x,f(x))
    
```

b

```python
fig, axes = plt.subplots(1,2,figsize=(20,5))
axes[1].set(xlabel='iter', ylabel='error',title=f"Iterative Descent")
axes[1].plot(range(len(xp)),ep,lw=1,marker='o',markersize=10)
axes[1].grid()
axes[1].set_yscale('log')

axes[0].set(xlabel='iter', ylabel='X',title=f"X evolution")
axes[0].plot(range(len(xp)),xp,lw=1,marker='o',markersize=10)
axes[0].grid()
axes[0].hlines([1],0,len(xp),lw=3,color='r')
axes[0].set_ylim(-1,10)



plt.show()
```
