# Monte Carlo alg for PI Calculation


```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np

print ("Libs random, math and pyplot loaded")
```

Lets generata a  problem, kind of DNA, with letter ABCDEF and a large sequence of, say, 1 M letters

```python
def norm(x,y): return math.sqrt(x*x + y*y)
def MC_PI(n) -> float:
    return 4*sum( True for _ in range(n) if norm(random.random(),random.random()) <= 1.0)/ n
MC_PI(10000000)
```

Lets see how good it works...

## Lets visualize

```python
fig, ax = plt.subplots(figsize=(10,10))

ns = 10000
ax.set(xlabel='x', ylabel='y',title=f"MonteCarlos Pi Estimation {ns} points")
ax.set_xlim( -0,1)
ax.set_ylim( -0,1)


for i in range(ns):
       x,y = (random.random(),random.random())   
       if norm(x,y) <= 1.0:
              color = 'r'
       else:
              color = 'b'
       ax.plot(x,y, marker='o', color= color,markersize=2)
       
       
# lets draw a simple arc
ts = np.arange(0,math.pi/2,math.pi/200, dtype=float)
x = [math.cos(t) for t in ts]
y = [math.sin(t) for t in ts]
ax.plot(x,y,lw=3,color='gray', linestyle='dotted')
       
       
plt.show()
```

## Precision increases as the number of samples increases

```python
n_s = list(range(1,10000))
pi_vals =  [MC_PI(i) for i in n_s]


# Data for plotting

fig, ax = plt.subplots()
ax.plot(n_s,pi_vals)

ax.set(xlabel='Number of samples)', ylabel='$pi$ estimate',
       title='MonteCarlos Pi Estimation')
ax.grid()

plt.show()
print (f"Mean of last 100 points is {sum(pi_vals[-100:])/100:.5f}")
```
